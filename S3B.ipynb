```
dataset = pd.read_csv("50_Startups.csv")
dataset.head()
```

```
dataset.isnull().any()
```

```
sns.relplot(x="R&D Spend", y="Profit", data = dataset)
```

```
sns.relplot(x="R&D Spend", y="Administration", data = dataset)
```

```
sns.relplot(x="R&D Spend", y="Marketing Spend", data = dataset)
```

```
sns.catplot(x="State", y="Profit", kind="bar", data=dataset)
```

```
sns.catplot(x="State", y="Profit", kind="box", data=dataset)
```

```
corr = dataset.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corr, annot=True, ax=ax)
```

```
sns.relplot(x="R&D Spend", y="Administration", hue= "State", data = dataset)
```

```
sns.relplot(x="Marketing Spend", y="Administration", hue= "State", data = dataset)
```
